```python
import os 
import pandas as pd
import numpy as np
import json
from tqdm import tqdm

import warnings
warnings.filterwarnings('ignore')
```

```python
base_directory = './Dataset/data/'
```

```python
for foldername, subfolders, filenames in tqdm(os.walk(base_directory), desc='Processing JSON Files', unit='file'):
    print(f"Current Subfolder: {foldername}")
    for filename in filenames:
        if filename == 'pm_data.json':
            print(f"Metadata File: {os.path.join(foldername, filename)}")
        else:
            print(f'Another File Present: {os.path.join(foldername, filename)}')
```

```python
import json

data = []
with open('./Dataset/data/28671856/pm_data.json', 'r') as file:
    for line in file:
        try:
            json_data = json.loads(line)
            data.append(json_data)
        except json.JSONDecodeError as e:
            print(f"Error decoding JSON: {e}")
            continue

# Now 'data' contains the valid JSON objects, and you can work with them as needed.
```

```python
# Accessing the first JSON object
data
```

```python
dfs = []

for foldername, subfolders, filenames in tqdm(os.walk(base_directory), desc='Processing JSON Files', unit='file'):
    for filename in filenames:
        if filename == 'pm_data.json':
            # Construct the full file path
            file_path = os.path.join(foldername, filename)
            # Read JSON data into a pandas DataFrame
            with open(file_path, 'r') as file:
                json_data = json.load(file)       
            with open(file_path, 'r') as file:
                try:
                    json_data = json.load(file)
                    df = pd.DataFrame(json_data)
                    dfs.append(df)
                except Exception as e:
                    print(f"Error processing file {file_path}: {e}")

# # Concatenate all valid DataFrames into a master DataFrame
if dfs:
    master_df = pd.concat(dfs, ignore_index=True)
    # Fill NaN values with null or any other desired value
    master_df = master_df.where(pd.notna(master_df), None)
    # Now you can perform operations on master_df
else:
    print("No valid JSON data found in the specified directory.")
```
